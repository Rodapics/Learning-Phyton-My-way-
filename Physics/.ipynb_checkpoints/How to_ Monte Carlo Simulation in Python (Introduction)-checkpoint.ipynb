#Introducción
El propósito de Monte Carlo es simular un experimento de la vida real que depende de la aleatoridad. Algunos ejemplos:

*   Contar le número de de fuentes de decaímiento radioactivos en un intervalo de tiempo $Δt$. La variable aleatoria aqui es el número de decaimiento N (entero positivo) y este sigue un modelo de probabilidad de **Poisson**:
$$f(x)=\frac{e^{-Λ}Λ^x}{x!} $$

*   Determinar la energía de las partículas en un detector de celdad en el ATLAS calorimeter en un intervalor de tiemp $Δt$. La variable aleatoria aqui es la energía **E** (númeor real y positivo) y la correspondiente función de densidad de probabilidad:
$$f(x)=F_1e^{-\sqrt{\frac{x}{E_1}}}+F_2e^{-\sqrt{\frac{x}{E_2}}} $$
*  **Estimar el valor de $π$**
