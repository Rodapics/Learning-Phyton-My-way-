```python
### Primero realizaremoss la conversión de cooddernadas esféricas a cartesianas.

def Sheperical2Cartesian(r, theta, phi):
  x = r*sin(theta)*cos(phi)
  y = r*sin(theta)*cos(phi)
  z = r*cos(theta)

  return x, y, z
```

```python
### Calcularemos los polinomios de Legendre.


def Legendre_Poly(n,x):

  if n== 0:
    P=1

  elif n==1:
    P=x

  else:
    P=(1/n)*((2*n-1)*x*Legendre_Poly(n-1,x)-(n-1)*Legendre_Poly(n-2,x))

  return P
```

```python
### Calcularemos los polinomios asociados de Legendre.


def Asso_Leg(l, m, x):
  try:
      if m == 0:
          P= Legendre_Poly(l,x)

      elif m > 0:
          P= (1/(sqrt(1-x**2)))*((l-m+1)*x*Asso_Leg(l,m-1,x)/ -(l+m-1)*Asso_Leg(l-1,m-1,x))
      
      elif m < 0:
          m = abs(m)
          P =((-1)**m)*(factorial(l-m)/factorial(l+m))*Asso_Leg(l, m, x)

  except ZeroDivisionError:
      P=0

  return P
```

```python
### Código Principal

l = int(input('Azimuthal quantum number "l": '))

if l < 0:
    print('"l" can not be a negative number')

m = int(input('Magnetic quantum number "m": '))

if m > l or m < (-l):
    print('quntum number "m" belong to "[-l,l]"')
```

```python

#%%  Normalization constant

K = sqrt( ((2*l + 1)* factorial(l - abs(m)))/ (4* pi* factorial(l + abs(m))) )

#%% Value of Phi and Theta

phi = linspace(0, 2* pi, 181)
tht = linspace(0, pi, 91)

Phi, Tht = meshgrid(phi, tht)
```

```python
#%% Value of Y

p, q = shape(Phi)
Y    = zeros([p, q])

for i in range(0, p):
    for j in range(0, q):

        if m > 0:
        
            Y[i, j] = sqrt(2) * K * cos(m * Phi[i, j]) * Asso_Leg(l, m, cos(Tht[i, j]))
        
        elif m < 0:
        
            Y[i, j] = sqrt(2) * K * sin(abs(m) * Phi[i, j]) * Asso_Leg(l, abs(m), cos(Tht[i, j]))
        
        else:
        
            Y[i, j] = K * Asso_Leg(l, 0, cos(Tht[i, j]))
```

```python
##%% Take care about negative Y
p, q = shape(Y)

#%% Finally convert data to cartesian form

x, y, z = Sheperical2Cartesian(abs(Y), Tht, Phi)
```
