#Quatum Mechanics Problem 1

Un electron está inicialmente preparado en un estado mecáno cuántico con momento angular nulo y función de Onda dado por:
$$Ψ(x,t=0)∝ exp[\frac{-x^2}{2a^2}] $$

El potencial en todas partes es cero. Un detector se ubica a una distancia $X=L$ de la posición inicial del eléctron. (Donde $L>>a$)



2.   Give the probability of observing the electron past $x=L$ as a function of time.

1.   Después de cuánto tiempo es probable que el electrón atraviese el detector?

2.   Dar la probabilidad de observar el electrón pasar por $X=L$ como una función del tiempo.



```python
#Librerias
import numpy as np
import sympy as smp
from sympy.functions import conjugate
from scipy.integrate import nquad
import matplotlib.pyplot as plt
```

Por simplicidad tenemos que $\hbar =1$

Lo primero que se puede realizar es graficar la función de Onda, para cualquier valor de a, solo buscamos observar su comportamiento en el tiempo.


```python
#Grafica
# a=1
x =np.linspace(-10,10,10000)
y= np.e**(-(x**2)/(2))
plt.plot(x,y)
plt.show()
```

Se observa un comportamiento Gaussiano.

##Solución Parte 1

  The solution of $⟨p^2⟩$ can be obtain from the wavefunction and using the classical stimation of $p$:
  $$p =m v= m\frac{L}{t} $$
  with t the time to get to the detector. Thus,

  $$ t=\frac{mL}{\sqrt{⟨p^2⟩}}$$ 

  

Por lo que, necesitamos calcular $\langle p^2\rangle$, de la siguiente manera:
$$⟨p^2⟩=\int_x  dx \psi \hat{p^2} ψ = \int_{-∞}^{∞} ψ (\frac{-d^2 \psi}{dx^2}) dx$$

```python
#Calculamos lo anterior de manera númerica.

x = smp.symbols('x', real=True)
a, L, m, k, kp, t = smp.symbols('a L m k k\' t', real=True, positive=True)
```

Now we define the normalize wave function.

```python
psi = smp.exp(-x**2 / (2*a**2))  #Define la función
psi = psi/smp.sqrt(smp.integrate(psi**2, (x, -smp.oo, smp.oo))) #Normaliza la función
```

```python
psi
```

Compute the value of $⟨p^2⟩$

```python
#Calculamos:
p2 = smp.integrate(psi*(-smp.diff(psi, (x,2))), (x,-smp.oo,smp.oo))
p2
```

Now we compute the time t given the equation above:

```python
T = m*L / smp.sqrt(p2)
T
```
