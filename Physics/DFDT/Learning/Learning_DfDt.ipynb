```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_1 = np.linspace(0,1,10)
x_1 = np.linspace(0,1,10)
X, T = np.meshgrid(x_1, t_1)
Y_1 = np.sin(np.pi * X) * np.cos(np.pi * T)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

color_map = ax.pcolormesh(T, X, Y_1, shading='auto', cmap='viridis')

cbar = fig.colorbar(color_map, ax=ax)

#cbar.set_label('y(x,t) = sin(πx) sin(πt)')  # Etiqueta de la barra de colores

# Etiquetas y título

ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Exact Solution')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la malla
dim = 10
t_2 = np.linspace(0, 1, dim)
x_2 = np.linspace(0, 1, dim)
Y_2 = np.zeros((dim, dim))
X_2, T_2 = np.meshgrid(x_2, t_2)

# Condición inicial (onda senoidal)
Y_2[:, 0] = np.sin(np.pi * x_2)

# Primer paso temporal usando una aproximación inicial
Y_2[1:dim-1, 0] = 0.5 * (Y_2[0:dim-2, 0] + Y_2[2:dim, 0])

# Iteración en el tiempo usando diferencias finitas
for n in range(1, dim-1):  # Tiempo
    for j in range(1, dim-1):  # Espacio
        Y_2[j, n+1] = 2 * Y_2[j, n] - Y_2[j, n-1] + 0.5 * (Y_2[j+1, n] - 2 * Y_2[j, n] + Y_2[j-1, n])

# Graficación del resultado
fig_2, ax_2 = plt.subplots(figsize=(8, 6))
color_map = ax_2.pcolormesh(T_2, X_2, Y_2, shading='auto', cmap='viridis')
cbar = fig_2.colorbar(color_map, ax=ax_2)

# Etiquetas y título
ax_2.set_xlabel('t')
ax_2.set_ylabel('x')
ax_2.set_title('Solución Numérica')

plt.show()
```
