##Matrices unidimensionales - de manera vertical


```python
#En forma de arreglo. Una matriz es un arreglo, un array.
#Rango 6

for i in range(1,7):
  print(i)

print('\n')
```

## Matrices unidimencionales - Horizontales

```python
#Rango 6

for i in range(1,7):
  print(i, end='')

print('\n')
```

##Matrices Bidimensionales
En este caso creamos matrices de $6*6$

```python
#Matriz de ceros
#Ciclo for anidado
for i in range(1,7):
  for j in range(1,7):
    print(0,end =' ')
  print('')
```

```python
#Matriz de ceros
#Ciclo for anidado
for i in range(1,7):
  for j in range(1,7):
    print(f'({i},{j})',end =' ')
  print('')
```

## Ejercicio. 
Colocar 1 en las diagonales


```python
#Matriz de ceros
#Ciclo for anidado
for i in range(1,7):
  for j in range(1,7):
    if (i==j):
      print(1,end =' ')
    else:
      print(0,end=' ')
  print('')
```
