Simular Onda Cuadrada.

 Que se representa de la siguiente forma:

$$f(x)=\frac{4}{π}\sum_{n=1,3,5...}^{∞}\frac{1}{n}sin(\frac{n\pi x}{L})$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal  #Crear Onda cuadrada base.
from celluloid import Camera  
```

```python
plt.style.use('seaborn')
```

```python
#Definir funcion fourier en python Onda cuadrada.

def fourier_onda_cuadrada(x,n):
  m = 2*n-1
  f = (4/np.pi)*(1/m)*np.sin(m*np.pi*x/L)
  return f

L = np.pi
ciclos = 2
x =np.linspace(-0.0001,2*ciclos*L,10000) #Negativo para ver bien la onda cuadrada


f = 0
n = 1
n_total = 3


#Aquí, realizamos la sumatoria.
#Las iteraciones las ponemos a mano
while n<n_total:
  f+=fourier_onda_cuadrada(x,n)
  plt.plot(x,f, label="n={}".format(2*n-1))
  n += 1


plt.plot(x, signal.square(x), color='k')
plt.legend()
plt.show()
```

Simular Onda en forma de rampa.
La ecuación caracteristica de esta onda es:

$$ f(x)= \frac{A}{2}-\frac{A}{\pi} \sum_{n=1}^{∞}(-1)^n*\frac{sin(\pi nx/L)}{n} $$

```python
def fourier_onda_cuadrada(x,n):
  f = (A/2)-((A/np.pi)*(-1)**n*np.sin(np.pi*n*x)/(n))
  return f

L = np.pi
A = 4
ciclos = 3
x =np.linspace(0,2*ciclos*L,10000) #Negativo para ver bien la onda cuadrada


f = 0
n = 1
n_total = 4



#Aquí, realizamos la sumatoria.
#Las iteraciones las ponemos a mano
while n<n_total:
  f+=fourier_onda_cuadrada(x,n)
  plt.plot(x,f, label="n={}".format(2*n-1))
  n += 1


plt.plot(x, signal.sawtooth(x), color='k')
plt.legend()
plt.show()
```
