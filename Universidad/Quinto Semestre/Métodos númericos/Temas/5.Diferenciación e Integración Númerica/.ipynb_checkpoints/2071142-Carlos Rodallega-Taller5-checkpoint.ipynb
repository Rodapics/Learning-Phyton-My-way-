###Sobre la teoria.
**Diferenciación.**
Recordemos la definición de derivada
$$f'(x_0)=\lim_{h→0} \frac{f(x_0+h)-f(x_0)}{h} $$




image.png

Tenemos varias maneras de aproximar a la derivada, ya sea de manera por delante o por detrás.

**Forward**:
$$f'(x_0)=\lim_{h→0} \frac{f(x_0+h)-f(x_0)}{h}  $$

**Backward**:
$$f'(x_0)=\lim_{h→0} \frac{f(x_0)-f(x_0+h)}{h}  $$

**Centrada**:
$$f'(x_0)=\lim_{h→0} \frac{f(x_0+h)-f(x_0-h)}{2h}  $$

De aquí, podemos decir que la aproximación hacía adelante:
$$ f'(x_0)≈\frac{f(x_1)-f(x_0)}{h}$$

Y podemos concluir:
$$f(x_{i+1}=f(x_i)+hf'(f(xi),xi)) $$

Lo anterior se conoce como el **Método de Euler**.

###Solución Punto a



Primero debemos identificar el tipo de EDO con la cual estamos trabjando, para encontrar su solución de manera más adecuada.

$$x'-x= sin(t) $$

Multiplicamos por $e^{-t}$ a ambos lados

$$e^{-t}x'-e^{-t} x= e^{-t}sin(t) $$

$$ ⇒ \frac{d}{dt} (e^{-t} x) = e^{-t}sin(t)$$

$$⇒  \frac{d}{dt} (e^{-t} x) dt = e^{-t}sin(t) dt$$

Integrando a ambos lados:

$$ (e^{-t} x)= -\frac{1}{2} e^{-t} (cos(t)+ sin(t))+ c_1$$

$$ ⇒ x(t)= -\frac{cos(t)}{2} -\frac{sin(t)}{2} + e^{t} c_1 $$

Dado que x(4)= 0.9

$$x(t)= -\frac{cos(t)}{2} -\frac{sin(t)}{2} + e^{t} (\frac{1.8+ cos(4)+ sin(4)}{2 e^{4}})$$



###Solución Punto b
---



```python
def real(x):
  return -1/2* np.cos(x) -1/2*np.sin(x) + np.exp(x)*((1.8+ np.cos(4)+ np.sin(4))/(2*np.exp(4)))
```

```python
#print("x(5.5)= ", real(5.5))
```

```python
plt.style.use('classic')
h= np.linspace(4, 5.5, 50)
plt.plot(h, real(h))
plt.xlabel("t")
plt.ylabel("x")
plt.title("Solución real")
plt.grid()
plt.show()
```

###Solución Punto a.




```python
#Calcular error como en el 2do taller.
def f(x,t):
  return x+ np.sin(t)


def euler(f,y0, x0, xf, n, real):
  h= (xf- x0)/n

  x1= np.zeros(n+1)
  y1= np.zeros(n+1)
  error= np.zeros(n+1)

  x1[0]= x0
  y1[0]= y0
  error[0]= np.abs(real(x0)-y0)

  for i in range(n):
    y1[i+1]= y1[i]+h* f(y1[i], x1[i])
    x1[i+1]= x1[i]+h
    error[i+1]= np.abs(real(x1[i+1])- y1[i+1])
  
  return x1, y1, error
```

```python
k=[10, 50, 100]
for i in k:
  print("Para k= " +str(i), " \n","Lista de x: ", euler(f, 0.9, 4, 5.5, i, real)[0]," \n ", "\n", "Lista de y: ", euler(f, 0.9, 4, 5.5, i, real)[1], " \n", " \n"," \n")
 
```

```python
for i in k:
  print("La solución para k= " +str(i), " es ", euler(f, 0.9, 4, 5.5, i, real)[1][i])
```

```python
for i in k:
  print("Errores para k= " +str(i), ":", euler(f, 0.9, 4, 5.5, i, real)[2], " \n")
```

### Solución Punto b.
Realizamos un procedimiento similar al realizado en el punto 1.

```python
def runge (f,y0, x0, xf, n, real):
  h= (xf- x0)/n

  x= np.zeros(n+1)
  y= np.zeros(n+1)
  error= np.zeros(n+1)

  x[0]= x0
  y[0]= y0

  error[0]= np.abs(real(x0)-y0)

  for i in range(n):
    k= f(y[i], x[i])
    k2= f(y[i]+ 0.5*h*k , x[i]+ 0.5*h)
    y[i+1]= y[i]+ h* ((k+k2)/2)
    x[i+1]= x[i]+h
    error[i+1]= np.abs(real(x[i+1])- y[i+1])
  
  return x, y, error
```

```python
k=[10, 50, 100]
for i in k:
  print("Para k= " +str(i), " \n","Lista de x: ", runge(f, 0.9, 4, 5.5, i, real)[0]," \n", " \n","Lista de y: ", runge(f, 0.9, 4, 5.5, i, real)[1], " \n", " \n", " \n")
  
```

```python
for i in k:
  print("La solución para k= " +str(i)," es ", runge(f, 0.9, 4, 5.5, i, real)[1][i])
```

```python
for i in k:
  print("Errores para k= " +str(i), ":", runge(f, 0.9, 4, 5.5, i, real)[2], " \n")
```

### Solución Punto c.





```python
def runge4 (f,y0, x0, xf, n,real):
  h= (xf-x0)/n
  
  x= np.zeros(n+1)
  y= np.zeros(n+1)
  error= np.zeros(n+1)

  x[0]= x0
  y[0]= y0

  for i in range(n):
    k= f(y[i], x[i])
    k2= f(y[i]+ 0.5*h*k , x[i]+ 0.5*h)
    k3= f(y[i]+ 0.5*h*k2 , x[i]+ 0.5*h)
    k4= f(y[i]+ h*k3 , x[i]+ h)
    
    y[i+1]= y[i]+ h* ((k+ 2*k2+ 2*k3+ k4)/6)
    x[i+1]= x[i]+h
    error[i+1]= np.abs(real(x[i+1])- y[i+1])
  
  return x, y, error
```

```python
k=[10, 50, 100]
for i in k:
  print("Para k= " +str(i), " \n","Lista de x: ", runge4(f, 0.9, 4, 5.5, i, real)[0]," \n", " \n","Lista de y: ", runge4(f, 0.9, 4, 5.5, i, real)[1], " \n", " \n", " \n")
```

```python
for i in k:
  print("La solución para k= " +str(i)," es ", runge4(f, 0.9, 4, 5.5, i, real)[1][i])
```

```python
for i in k:
  print("Errores para k= " +str(i), ":", runge4(f, 0.9, 4, 5.5, i, real)[2], " \n")
```

**Comentarios**:

Es notorio a simple vista que entre mayor sea el número de pasos, mayor sera el acercamiento de nuestro método a la solución real. Esto tiene explicación teorica, debido a que entre mayores pasos tengamos, nuestro 'dx'(h) interno de la función se hará cada vez más pequeño, realizando la analogía con la definición de derivada, cuando vemos que nuestro 'dx' tiende a cero, estamos en el valor exacto de nuestra función. Así, con pasos mayores, nuestro 'dx' tiende a cero y se obtendrá una aproximación más exacta y precisa.

De la misma manera, el error relativo debe disminuir entre más pasos usemos, como habíamos mencionado, más pequeño será el 'dx' y mejor será la , esto lo podemos notar en nuestra segunda grafica que para valores muy pequeños del pasos, obtenemos un error bastante grande, mientras, a valores grandes el error se mantiene pequeño. Vemos que el error se hace cada vez más grande a medida que aumentamos el número de pasos, esto es debido a que el error obtenido en cada paso se mantiene, nunca estaremos en un paso con la solución original y esto ocaciona que a medida que transcurre nuestra función en su dominio, el error se irá haciendo más grande.
