#Diferenciación.
Recirdemos la definición de derivada
$$f'(x_0)=\lim_{h→0} \frac{f(x_0+h)-f(x_0)}{h} $$




image.png

Tenemos varias maneras de aproximar a la derivada, ya sea de manera por delante o por detrás.

**Forward**:
$$f'(x_0)=\lim_{h→0} \frac{f(x_0+h)-f(x_0)}{h}  $$

**Backward**:
$$f'(x_0)=\lim_{h→0} \frac{f(x_0)-f(x_0+h)}{h}  $$

**Centrada**:
$$f'(x_0)=\lim_{h→0} \frac{f(x_0+h)-f(x_0-h)}{2h}  $$

De aquí, podemos decir que la aproximación hacía adelante:
$$ f'(x_0)≈\frac{f(x_1)-f(x_0)}{h}$$

Y podemos concluir:
$$f(x_{i+1}=f(x_i)+hf'(f(xi),xi)) $$

Lo anterior se conoce como el **Método de Euler**.

#Método de Euler

```python
def euler(f,x0,xf,y0,h):
  n = int( np.ceil( (xf -x0)/h ))
  x= np.zeros(n)
  y= np.zeros(n)
  x[0]= x0
  y[0]= y0
  for i in range(n-1):
    y[i+1]= y[i]+h* f(x[i],y[i])
    x[i+1]= x[i]+h 
  return x,y
```

```python
#Funciones para probar código.
def f(x,y):
  return np.cos(x)


def real(x):
  return np.sin(x) + 2

t=np.linspace(0,10,1000)
y=np.sin(t)+1
```

```python
#Como la funcion derivada es un coseno, la función aproximada encontrada debe ser un seno.
exsin,eysin = euler(f,0,10,1,0.1)
```

```python
#Gráficación de los resultados:
plt.style.use('classic')
plt.plot(t,y,label='Coseno real')
plt.plot(exsin,eysin,'o' ,label='Coseno Euler')
plt.xticks([0,np.pi/2,np.pi,np.pi*3/2,np.pi*2,np.pi*5/2],[r'$0$',r'$\pi/2$',r'$\pi$',r'$3/2\pi$',r'$2\pi$',r'$5/2\pi$']) #Colcar los nombres de los Pi's.
plt.legend(loc='upper left')
plt.xlabel("Dominio")
plt.ylabel("Rango")
plt.xlim(0,10)
plt.ylim(0,2.5)
plt.title('Función para un paso fijo')
plt.grid()
plt.show()
```
