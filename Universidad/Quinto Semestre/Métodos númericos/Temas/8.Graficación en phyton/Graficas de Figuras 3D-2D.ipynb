\Ejemplo 11.15 - Gr√°fica del campo man√©tico de un alambre usando quiver

```python
#Pre trabajo
from pylab import*
xmax = 2.0
xmin = -xmax
NX = 5
ymax = 2.0
ymin = -ymax
NY = 5
```

```python
#Componentes vectoriales y puntos.
x = np.linspace(xmin, xmax, NX)
x = np.linspace(ymin, ymax, NY)
x, y = meshgrid(x, y)
S2 = X**2 + Y**2  #Este es el radio al cuadrado
```

```python
#Campo magn√©tico. El valor de 0.001 es para evitar divisi√≥n entre cero.
Bx = -Y/(S2+0.001)
By = +X/(S2+0.001)
```

```python
fig = plt.figure(figsize=(10,8))
QP = plt.quiver(X, Y, Bx, By)
quiverkey(QP,0.85,0.85,1.0,"Campo magn√©tico")
#L√≠mites de los ejes
dx = (xmax-xmin)/(NX-1.)
dy = (ymax-ymin)/(NY-1.)
plt.axis([xmin-dx,xmax+dx,ymin-dy,ymax+dy])
plt.title("Campo magn√©tico de un alambre de con corriente")
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.savefig("Grafica_Ejemplo_11_19.png",dpi=300)
plt.show()
```

Ejemplo 11.16- Gr√°fica de superficie con proyecci√≥n sobre el plano x,y.

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(10,8))
ax = Axes3D(fig)
X = np.arange(-4,4, 0.25)
Y = np.arange(-4,4, 0.25)
X,Y = np.meshgrid(X, Y)
R = np.sqrt(abs(X**3)+Y**2)
Z = np.sin(R)

A = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis') 
ax.contourf(X, Y, Z, zdir="z", offset=-2, cmap='viridis')
ax.set_zlim(-2,2)
fig.colorbar(A, shrink=0.5, aspect=5)
plt.title("Pinche grafica fea")
plt.savefig("Grafica_Ejemplo_11_20.png", dpi=300)
plt.show()
```

Ejemplo de una funci√≥n Parametrica en 2D

```python
#Grafica de una funci√≥n Parametrica en phyton (2D) No continua
import numpy as np
import pylab as plt
import math
import sympy as sp
from sympy import *

t = symbols('t')

 
xexp = '3*t'  #exep, expresi√≥n
yexp = 't-5'

#Cambiar las expresiones de variable t a numeritos para poder evaluar.
x = sp.lambdify((t),xexp,"numpy")
y = sp.lambdify((t),yexp,"numpy")

#print(x(1))   #Evaluando.
#print(y(1))

#Crear espacio de la variable t
t = np.linspace(0,25,10, endpoint=True)   #Endopoint, tomar el √∫ltimo valor.
#print(t)



#Evaluar las expresiones en el parametro t
vecx = x(t)
vecy = y(t)
#print(vecx)
#print(vecy)


#Graficar

plt.plot(vecx,vecy,'o')
plt.title("Grafica Parametrica")
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.savefig("Grafica_Parametrica.png", dpi=1000)
plt.show()
```

Grafica Parametrica. Problema de la braquistocrona.

```python
#Grafica parametrica del problema de la braquistocrona
#Librerias
import numpy as np
import pylab as plt
import math
import sympy as sp
from sympy import *
#Simbolos.

t = symbols('t')

#Expresiones 

xexp = '1*(t-sin(t))'  #exep, expresi√≥n
yexp = '-1*(1-cos(t))+2'

#Cambiar las expresiones de variable t a numeritos para poder evaluar.
x = sp.lambdify((t),xexp,"numpy")
y = sp.lambdify((t),yexp,"numpy")

#print(x(1))   #Evaluando.
#print(y(1))

#Crear espacio de la variable t
t = np.linspace(0,1*math.pi,30, endpoint=True)   #Endopoint, tomar el √∫ltimo valor.
#print(t)

#Evaluar las expresiones en el parametro t
vecx = x(t)
vecy = y(t)
#print(vecx)
#print(vecy)


#Graficar

plt.plot(vecx,vecy,'-')
plt.title("Gr√°fica de la Braquistocrona")
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.savefig("Grafica_Parametrica.png", dpi=1000)
plt.show()
```
