Escriba un programa en python para calcular la constante matematica $e$, a partir de la definición $$ e = \lim_{n \to \infty }(1+\frac{1}{n})^{n}$$, es decir , calcula $ (1+\frac{1}{n})^{n}$ para $n=10^{k}  , k= 1,2,..,20$ .Determina el error en tus aproximaciones comparándolas con $exp(1)$ (Usa la función $exp$ de Numpy), genera un gráfico para visualizar los resultados(Comenta y analiza tus resultados).  

```python
#Librerias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Solución

aprox = [0]

#Ciclo para calcular el valor aproximado
for k in range(1,20,1):
  n=10**k
  e_exp = (1+(1/n))**n
  aprox.append(e_exp)
print('Valor aproximado obtenido:',aprox)


#Continuación definición de e por libreria y calculo de errores
e = np.exp(1)
Errores = abs(aprox-e)
print('Error obtenido:',Errores)

#Valor de n en un array para graficar
n = [0]
for k in range(0,19,1):
  n_1=10**k
  n.append(n_1)
nx=np.array(n)
```

```python
#Grafica Errores vs n
##Tema

plt.loglog(nx,Errores,'-o', label='Línea de tendencia')
plt.legend()
plt.title('Errores al calcular el número e computacionalmente')
plt.xlabel('Valor de n')
plt.ylabel('Errores')
plt.grid()
plt.show()
```

### Análisis de resultados. Comentarios.
Como podemos apreciar visualmente en la imagen anterior, el error de cálculo del número $e$, decrece linealmente al aumentar el valor de $h$, esto se debe a que el error de redondeo se disminuye al tener valores de $n$ demasiado grandes podemos hacer tender $e$ más al valor del límite: $$ e = \lim_{n \to \infty }(1+\frac{1}{n})^{n}$$ Lo ideal, es que al seguir tendiendo $n→∞$ el error de redondeo siga reduciendose de manera lineal.
Como también podemos observar, esto no es lo que sucede, para n entre $10^6$ y $10^8$ podemos observar que el error deja de decrecer y empieza a crecer. Esto es debido al error de truncamiento se hace envidente, este error es inherente a los computadores, debido a que no podemos representar números infinitos en nuestros computadores, o nuestro computador no los puede almacenar en la mantiza, vemos que el error total no esta solo compuesto del error de redondeo, sino, se compone de la suma del error de redondeo y el de truncamiento. Pero, ¿por qué a esos valores de n empezamos a encontrar este error de truncamiento? ¿Por qué no para $n=10^2$ ó $n=10^50$?

Esto se debe a la composición básica de nuestros computadores. Especificamente se debe a la cantidad de bits que usamos para almacenar los números en la computadora. Al tener una cantidad de bits finitos, tenemos un número finito de números. Separados por cierto intervalo conocido como epsilón de máquina. Cuando tenemos operaciones que tratan cantidades menores a este epsilón de máquina, el computador interpreta estos valores como un 0, y por eso, todos los avances en el cálculo del número $e$ (avances reduciendo el error con el redondeo) se empieza a desperdiciar cuando el computador se ve afectado por los errores de truncamiento en aproximadamente n=$10^7$.

Podemos observar que no tiene ningún sentido ni podemos obtener un cálculo preciso de algo cuando tenemos valores extremadamente grande de n, vemos que el error se hace demasiado evidente. Hay que tener prudencia en los cálculos y las iteraciones.

Supongase que quieres calcular $n+1$ puntos equiespaciados en el intervalo [a,b], con espacio $h=(b-a)/n$.


1.   Usando arimética de punto flotante, cuál de los siguientes dos métodos es más preciso y por qué?

*   Método 1:
$$ x_0=a ,\:\:\:\:\:\:  x_k=x_{k-1}+h,\:\:\:\:\: k=1,2,3,...n$$
*   Método 2:
$$  x_k= a+kh, \:\:\:\:\: k=0,1,...,n$$


2.   Implementa ambos métodos con $ a=0$ y $b=1$ para ilustrar la diferencia, calcula las diferencias absolutas entre cada elementos de los espaciamientos y genera un grafico que permita visualizarlas (Comenta y analiza tus resultados).



```python
#Solución primer punto y segundo punto
#Primer método.
a=0
b=1
n=100
h=(b-a)/n
met1 = [0]
x_0=a
for k in range(1,101):
  x_k=x_0+h
  x_0=x_k
  met1.append(x_k)
print('Método 1:',met1)

mt1=np.array(met1)
#Segundo método
a=0
b=1
n=100
h=(b-a)/n
met2 =[0]
for k in range(1,101):
  x_k=a+k*h
  met2.append(x_k)
print('Método 2:',met2)
mt2=np.array(met2)
#Grafica
Dif = abs(mt1-mt2)
#print('Diferencia entre los dos métodos',Dif)
x1=np.arange(0,101,1)
#print('Error:',Dif)

```

```python
#Grafica

plt.plot(x1,Dif,'-',label='Línea de tendencia')
plt.title('Dif entre dos métodos computacionales')
plt.xlabel('n')
plt.legend()
plt.legend(loc='upper left')
plt.ylabel('Diferencia entre ambos métodos')
plt.grid()
plt.show()
```

###Análisis de resultados. Comentarios.

Podemos apreciar visualmente que a medida que el valor de n tiende a valores muy grandes, la diferencia entre los métodos se hace grande. Según el enunciado, esto aritméticamente no tiene ningún sentido, porque ambos métodos deberían tener una diferencia de 0.
Obviamente, eso es lo que encontramos de manera aritmética, pero cuando realizamos estos cálculos en computadora, debemos entender que tenemos límitaciones practicas y técnicas, similar al ejercicio anterior, para el primer método, el computador necesita utilizar gran espacio en la mantiza para realizar las operaciones. Mientras en el segundo método, ocupa menos espacio porque define operaciones de manera directa a través de la multiplicación, así, lo que se hacía iteradamente en la primera parte, se realiza de manera más directa en el segundo método, evitando coleccionar errores de truncamiento en cada cálculo.
