```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

Arr1 = pd.read_csv("Datos Inverso cuadrado.csv")
x = Arr1["x"]
y = Arr1["y"]

X= np.log(x)
Y= np.log(y)
def fun(x,m, b):
    return m*x+b


res, cov= curve_fit(fun, X, Y)


plt.plot(x, y, "ro-", label='Datos') 
plt.legend()
plt.grid(True)
#plt.title('Histograma y Ajuste Gaussiano')
plt.xlabel(r"(Distancia) ")
plt.ylabel(r'$\frac{N}{s}$')

plt.show()


r= np.arange(0, 3, 0.2)
plt.plot(r, fun(r, res[0], res[1]), 'r-', label='Ajuste')
plt.plot(X, Y, "bo", label='Datos') 
plt.legend()
plt.grid(True)
#plt.title('Histograma y Ajuste Gaussiano')
plt.xlabel(r"Ln(Distancia) ")
plt.ylabel(r'Ln($\frac{N}{s}$)')

plt.show()




print("Exponente = ", -1*res[0])
print("Parametro = ", -1*res[1])
```
