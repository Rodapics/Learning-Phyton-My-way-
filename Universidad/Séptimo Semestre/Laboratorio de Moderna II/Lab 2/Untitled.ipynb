```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

Arr1 = pd.read_csv("Datos Inverso cuadrado.csv")
x = Arr1["x"]
y = Arr1["y"]

X= np.log(x)
Y= np.log(y)
def fun(x,m, b):
    return m*x+b


res, cov= curve_fit(fun, X, Y)


plt.plot(x, y, "ro-", label='Datos') 
plt.legend()
plt.grid(True)
#plt.title('Histograma y Ajuste Gaussiano')
plt.xlabel(r"(Distancia) ")
plt.ylabel(r'$\frac{N}{s}$')

plt.show()
```
