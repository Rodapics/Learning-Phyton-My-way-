```python
from scipy.optimize import curve_fit


Arr1 = pd.read_csv("Datos N.csv")
N= Arr1["N"]

hist, bins = np.histogram(N, bins=18)
width = 0.8 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2


fig= plt.figure()


plt.bar(center, hist, align='center', width=width, label='Histograma')


def gaussian(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))


popt, pcov = curve_fit(gaussian, center, hist, p0=[1, np.mean(N), np.std(N)])

r= np.arange(3700, 4500,1)
plt.plot(r, gaussian(r, *popt), 'r-', label='Ajuste Gaussiano')
plt.legend()
plt.xlim(3700, 4500)
plt.grid(True)
plt.title('Histograma y Ajuste Gaussiano', fontsize = 20)
plt.xlabel('$N$', fontsize = 20)
plt.xlim(3800,4300)
plt.ylabel('Frecuencia', fontsize = 20)
plt.text(0.2,0.7,textstr1, transform=fig.transFigure, bbox=dict(facecolor='white', edgecolor='black'), fontsize= 15)
plt.savefig("Grafica_Ajuste_1.png", dpi=500)
plt.show()


print("Amplitud = ", popt[0])
print("Media = ", popt[1])
print("Desviación estandar = ", popt[2])

```
