```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots 
from scipy.constants import milli
from scipy.stats import linregress

```

```python
#Diseño de gráficas
plt.style.reload_library()
plt.style.use(['science','notebook'])
```

# Hierro

```python
xH,TH=np.loadtxt(fname="Datos Hierro.csv",dtype=float,delimiter=',',skiprows=1,usecols=(0,5),unpack=True)
```

```python
(xH,np.log(TH))
```

```python
plt.plot(xH,TH,'-o')

plt.xlabel('Espesor ($mm$)')
plt.ylabel('Partículas por segundo (N/s)')

plt.grid(linestyle='--')
plt.savefig(fname='Tvsx_Hierro.png',dpi=500)
```

```python
AdjustH=linregress(xH,np.log(TH))

uH=-AdjustH.slope
uH

#(np.log(I0),AdjustH.intercept)
```

```python
plt.plot(xH,np.log(TH),'-o')
plt.plot(xH,AdjustH.slope*xH+AdjustH.intercept,'--')

plt.xlabel('Espesor ($mm$)')
plt.ylabel('$Ln(N/s)$')
plt.legend(['$Ln(T)$ vs $x$','Ajuste Lineal'])

plt.grid(linestyle='--')
plt.savefig(fname='AjusteHierro.png',dpi=500)

plt.show()
```

```python
(AdjustH.rvalue)**2
```

```python
print(f'El coeficiente de atenuación lineal del Hierro es: {uH:.3f} +- {AdjustH.stderr:.3f} 1/mm')
print("La distancia media (d1/2) del hierro es: ", (np.log(2))/uH)
```

```python

#* Cálculo de x medio

xmH=np.abs(np.log(0.5)/-uH) 
dxmH=xmH*(AdjustH.stderr/uH)

print(f"El d_(1/2) Hierro es: {xmH:.2f} +- {dxmH:.2f} mm")
```
