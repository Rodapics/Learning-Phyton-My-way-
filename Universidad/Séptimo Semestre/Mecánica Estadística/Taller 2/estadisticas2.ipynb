```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sp
```

```python
N1=1
N2=1

m=1
k=sp.Boltzmann

#H varía
H1=np.linspace(1,100,100)
E1=1

#E varía
E2=np.linspace(0.0001,0.02,1000)
H2=1
```

```python
fig, ax1 = plt.subplots(dpi=500)

T1=(H1**2*N1**2*m**2 - E1**2)/(N1*k*E1)
T2=(H2**2*N2**2*m**2 - E2**2)/(N2*k*E2)

ax1.plot(H1, T1, 'b-')
ax1.set_xlabel('Intensidad magnética',color="b")
ax1.set_ylabel('Temperatura')
ax1.tick_params('x', colors='b')
ax1.set_xticks([0,20,40,60,80,100],[])

ax2 = ax1.twiny()
ax2.plot(E2, T2, 'r-')
ax2.set_xlabel('Energía', color='r')
ax2.tick_params('x', colors='r')
ax2.set_xticks([0,0.0025,0.005,0.0075,0.01,0.0125,0.015,0.0175,0.02],[])
ax1.set_yticks([0,1e26,2e26,3e26,4e26,5e26,6e26,7e26],[])
ax1.grid()
```
