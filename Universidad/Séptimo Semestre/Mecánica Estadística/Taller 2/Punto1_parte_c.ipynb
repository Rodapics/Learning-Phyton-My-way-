```python
#Desviación estándar. Calcula la desviación estandar de esos datos y le asigna el respectivo símbolo.
textstr1 = '$\epsilon = E/(\mu*B*N)$'
```

```python
#Graficar entropía esa

x = np.linspace(-1,1,1000)  #Creamos la variable de energía

N = 100 #número de partículas en el sistema

u = 1 # momento magnetico de las partículas

B = 1 #Magnitud del campo magnetico aplicado

S_k = -N*((((1-x)/2)*(np.log((1-x)/2)))+(((1+x)/2)*(np.log((1+x)/2))))
```

```python
fig= plt.figure()


plt.plot(x,S_k, label='Entropía')

plt.title('Entropía en función de la energía ($\epsilon$)',  fontsize = 20)
plt.ylabel('Entropía/$k_B$', fontsize = 20)
plt.xlabel('$ (\epsilon)$', fontsize = 20)
plt.legend()
plt.grid()
plt.text(0.45,0.2,textstr1, transform=fig.transFigure, bbox=dict(facecolor='white', edgecolor='black'), fontsize= 15)
plt.savefig('grafica1.png', dpi=300) 
plt.show()
```

```python
fig= plt.figure()


plt.plot(x,T, label='Temperatura')

plt.title('Temperatura en función de la energía ($\epsilon$)',  fontsize = 20)
plt.ylabel('Temperatura', fontsize = 20)
plt.xlabel('$ (\epsilon)$', fontsize = 20)
plt.ylim(-10,10)
plt.legend()
plt.grid()
plt.text(0.2,0.2,textstr1, transform=fig.transFigure, bbox=dict(facecolor='white', edgecolor='black'), fontsize= 15)
plt.savefig('grafica2.png', dpi=300) 
plt.show()
```
