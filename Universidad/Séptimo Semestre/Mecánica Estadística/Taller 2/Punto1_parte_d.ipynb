```python
x = np.linspace(0,100,1000)  #Creamos la variable de temperatura

e_b = np.exp(1/x)


u=1
b=10

e_ub = np.exp(1/(2*u*b))


#Definimos la energía

e = (e_b-e_ub)/(e_b-e_ub)
```
