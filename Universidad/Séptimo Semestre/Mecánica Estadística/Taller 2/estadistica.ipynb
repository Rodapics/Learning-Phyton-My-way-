```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sp
```

```python
def S(m,N,H,E):
    return N*sp.Boltzmann*(np.log(N)-np.log((N/2 - E/(2*m*H))**0.5) - np.log((N/2 + E/(2*m*H))**0.5))
```

```python
E=np.linspace(-100,0,100)
N=100
H=1
m=1
S=S(m,N,H,E)
```

```python
plt.figure(dpi=500)

plt.plot(E,S)
plt.xticks([0,-20,-40,-60,-80,-100],["0","","","","",""])
#plt.yticks([1e-21,1.5e-21,2e-21,2.5e-21,3e-21,3.5e-21],[])
plt.xlabel("Energía")
plt.ylabel("Entropía")
plt.grid()
plt.xlim(-105,0)
```
