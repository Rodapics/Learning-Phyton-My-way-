# Primera Parte 1: Distancia entre un par de partículas.
Sabemos que dos partículas pueden colisionar, y eso es lo más importante en este problema que vamos a resolver debido a que saber el estado después de colisionar es lo ensencial. Para determinar si dos partículas han colisionado necesitamos definir una **distancia** entre todas los posibles pares de partículas.
Ahora imagínese que tiene dos esferas, si estas se ubican en un espacio menor que dos veces el radio de cada esfera, dichas esferas se tocarán. El mismo caso sucede con las partículas, si la distancia entre partículas en mucho menor que dos veces el radio de cada partícula, ambas partículas chocarán.
Si tenemos $n$ partículas, entonces se tiene un total de 
$$\frac{n(n-1)}{2} $$
pares de combinaciones posibles.

```python
#list(combinations(ids,2)) #Combinations solo trabaja en forma de lista
# y me da las combinaciones posibles de los números que le doy (primera entrada)
# y me los combina en los números que yo quiero (2da entrada.)
```

```python
# para realizar esto, le colocaremos una etiqueta a cada par de partículas posible
#asarray nos permite convertir una lista en un arreglo.
ids_pairs = np.asarray(list(combinations(ids,2)))
ids_pairs
```

De una manera similar, podemos obtener las distancias de las partículas

```python
#Creamos los pares de posiciones de la partícula en el eje X.
#Estamos creando todos los posibles pares de posiciones para las partículas.(EJE X)

x_pairs= np.asarray(list(combinations(r[0],2)))
x_pairs
```

Ahora tomamos la diferencia entre cada par de partículas obtenemos: $$\Delta X_{ij} $$
que son los diferenciales de cada par de partículas.

```python
dx_pairs = np.diff(x_pairs,axis=1, n=1).ravel()
dx_pairs
```

La distancia total entre las partículas es simplemente 
$$d_{ij}=\sqrt(\Delta x_{ij}^2 + y_{ij}^2) $$
A su vez, como tenemos $n(n-1)/2$ pares, se tendrán $n(n-1)/2$ distancias, una para cada valor de $d_{ij}$.

```python
x_pairs= np.asarray(list(combinations(r[0],2)))
y_pairs= np.asarray(list(combinations(r[1],2)))
dx_pairs = np.diff(x_pairs,axis=1, n=1).ravel()
dy_pairs = np.diff(y_pairs,axis=1, n=1).ravel()
d_pairs = np.sqrt(dx_pairs**2+dy_pairs**2)
d_pairs
```

Vemos que así se puede obtener la distancia entre cada una de las partículas.

# Segunda parte: Velocidad de una colisión.
Como conocemos de la mecánica newtoniana, cuando las partículas colisionan, estas intercambian velocidad de tal manera que su momento líneal se conserva. Así, para cada iteración de la simulación, si hay una colisión, cada velocidad de dichas particulas que colisionan, ambas velocidades deben actualizarse.

Sabemos que la colisión sucede si la distancia entre las partículas es menor a dos de dos veces el radio de cada partícula. La velocidad final de cada una de las partículas estarán dadas por la siguiente relación:
$$\vec{v_1^{new}}= \vec{v_1} -\frac{(\vec{v_1}- \vec{v_2})\cdot(\vec{r_1}-\vec{r_2})}{|\vec{r_1}-\vec{r_2}|^2} (\vec{r_1}-\vec{r_2}) $$
$$\vec{v_2^{new}}= \vec{v_2} -\frac{(\vec{v_2}- \vec{v_1})\cdot(\vec{r_2}-\vec{r_1})}{|\vec{r_1}-\vec{r_2}|^2} (\vec{r_2}-\vec{r_1}) $$



```python
# Como hemos dicho, es importante definir el radio de las partículas
#Este es un valor fijo y será el mismo para cada una de las partículas.

radius = 0.06

#Hallamos las partículas que han colisionado:
ids_pairs_collide = ids_pairs[d_pairs < 2*radius]
ids_pairs_collide 
```

```python
#Esta función permite obtener un array de valores de verdad, que me indican si hubo colisión o no

d_pairs < 2*radius

#Al evaluar esta condición en los posibles pares de partículas, solo nos quedamos con
# los pares de partículas que han colisionado
```

Ahora ya podemos calcular la nueva velocidad. Lo más complicado de calcular es el producto punto entre los dos vectores:
$$(\vec{v_2}- \vec{v_1})\cdot(\vec{r_2}-\vec{r_1}) $$
Procedemos a calcular esto primero y después dejamos la expresión completa para la nueva velocidad.


```python
#Calculo del producto punto.
c_dot1 = np.diag((v1-v2).T @ (r1-r2))
c_dot2 = np.diag((v2-v1).T @ (r2-r1))
```

```python
#Ahora si, calculando la velocidad:
v1new = v1 - c_dot1/np.sum((r1-r2)**2, axis= 0) * (r1-r2)
v2new = v2 - c_dot2/np.sum((r2-r1)**2, axis= 0) * (r2-r1)
```

# Tercera Parte: Funciones necesarias para la simulación.
A continuación colocamos algunas funciones que nos facilitarán la simulación.

```python
def get_delta_pairs(x, ids_pairs):
    return np.diff(np.array([x[ids_pairs[:,0]], x[ids_pairs[:,1]]]).T, axis=1).ravel()

def get_deltad_pairs(r, ids_pairs):
    return np.sqrt(get_delta_pairs(r[0], ids_pairs)**2 + get_delta_pairs(r[1], ids_pairs)**2)

def compute_new_v(v1, v2, r1, r2):
    v1new = v1 - ((v1-v2)*(r1-r2)).sum(axis=0)/np.sum((r1-r2)**2, axis=0) * (r1-r2)
    v2new = v2 - ((v1-v2)*(r1-r2)).sum(axis=0)/np.sum((r2-r1)**2, axis=0) * (r2-r1)
    return v1new, v2new

def motion(r, v, id_pairs, ts, dt, d_cutoff):
    rs = np.zeros((ts, r.shape[0], r.shape[1]))
    vs = np.zeros((ts, v.shape[0], v.shape[1]))
    # Initial State
    rs[0] = r.copy()
    vs[0] = v.copy()
    for i in range(1,ts):
        ic = id_pairs[get_deltad_pairs(r, ids_pairs) < d_cutoff]
        v[:,ic[:,0]], v[:,ic[:,1]] = compute_new_v(v[:,ic[:,0]], v[:,ic[:,1]], r[:,ic[:,0]], r[:,ic[:,1]])
        
        v[0,r[0]>1] = -np.abs(v[0,r[0]>1])
        v[0,r[0]<0] = np.abs(v[0,r[0]<0])
        v[1,r[1]>1] = -np.abs(v[1,r[1]>1])
        v[1,r[1]<0] = np.abs(v[1,r[1]<0])
        
        r = r + v*dt
        rs[i] = r.copy()
        vs[i] = v.copy()
    return rs, vs
```

Fijamos el radio y encontramos la posición de todas las partículas en función del tiempo (usamos arreglos 3D donde es más fácil obtener $(t,x,y)$)

```python
radius = 0.03
rs, vs = motion(r, v, ids_pairs, ts=1000, dt=0.000008, d_cutoff=2*radius)
```
