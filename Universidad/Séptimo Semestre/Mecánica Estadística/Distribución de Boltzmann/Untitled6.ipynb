```python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros de la mol√©cula de ox√≠geno
m = 5.31e-26  # Masa de una mol√©cula de ox√≠geno en kg
T = 300       # Temperatura en Kelvin

# Funci√≥n de distribuci√≥n de velocidades de Maxwell-Boltzmann
def maxwell_boltzmann(v, m, T):
    return (np.sqrt(2 / np.pi) * np.power((m / (2 * np.pi * T)) , 3 / 2) * np.exp(-m * v**2 / (2 * T)))

# Rango de velocidades
v = np.linspace(0, 3000, 1000)  # Valores de velocidad en m/s

# Calcula la distribuci√≥n de velocidades de Maxwell-Boltzmann
f = maxwell_boltzmann(v, m, T)

# Graficar la distribuci√≥n de velocidades
plt.plot(v, f)
plt.xlabel('Velocidad (m/s)')
plt.ylabel('Distribuci√≥n de velocidades')
plt.title('Distribuci√≥n de velocidades de Maxwell-Boltzmann \n para una mol√©cula diat√≥mica de ox√≠geno a 300 K')
plt.grid(True)
plt.show()
```
