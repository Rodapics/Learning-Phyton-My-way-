```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
#Energ√≠a

#Primera Grafica----------------------------------------

fig= plt.figure(figsize=(20,15))

E= 10
k= 10

m=5


a = np.sqrt(2*E/k) #Eje mayor de la elipse

b = np.sqrt(2*m*E) #Eje menor de la elipse


t = np.linspace(0,360,360)

x =a*np.cos(np.radians(t)) #10 is minor axis of ellipse

y = b*np.sin(np.radians(t)) #50 is major axis of ellipse


plt.plot(x,y, label = 'Energ√≠a = $10$, Constante $K$ = $10$ ')
#plt.gca().set_aspect('equal')
plt.legend(fontsize = 20)
plt.grid()



#Segunda Grafica----------------------------------------
E= 20
k= 10

m=5


a = np.sqrt(2*E/k) #Eje mayor de la elipse

b = np.sqrt(2*m*E) #Eje menor de la elipse


t = np.linspace(0,360,360)

x =a*np.cos(np.radians(t)) #10 is minor axis of ellipse

y = b*np.sin(np.radians(t)) #50 is major axis of ellipse

plt.plot(x,y, label = 'Energ√≠a = $20$, Constante $K$ = $10$ ')
#plt.gca().set_aspect('equal')
plt.legend(fontsize = 20)
plt.grid()

#Tercera Grafica----------------------------------------

E= 30
k= 10

m=5


a = np.sqrt(2*E/k) #Eje mayor de la elipse

b = np.sqrt(2*m*E) #Eje menor de la elipse


t = np.linspace(0,360,360)

x =a*np.cos(np.radians(t)) #10 is minor axis of ellipse

y = b*np.sin(np.radians(t)) #50 is major axis of ellipse

plt.plot(x,y, label = 'Energ√≠a = $30$, Constante $K$ = $10$ ')
plt.legend(fontsize = 20)
plt.grid()

#Cuarta Grafica-----------------------------------
E= 10
k= 20

m=5


a = np.sqrt(2*E/k) #Eje mayor de la elipse

b = np.sqrt(2*m*E) #Eje menor de la elipse


t = np.linspace(0,360,360)

x =a*np.cos(np.radians(t)) #10 is minor axis of ellipse

y = b*np.sin(np.radians(t)) #50 is major axis of ellipse

plt.plot(x,y, label = 'Energ√≠a = $10$, Constante $K$ = $20$ ')
plt.legend(fontsize = 20)
plt.grid()

#Quita Grafica-----------------------
E= 10
k= 20

m=5


a = np.sqrt(2*E/k) #Eje mayor de la elipse

b = np.sqrt(2*m*E) #Eje menor de la elipse


t = np.linspace(0,360,360)

x =a*np.cos(np.radians(t)) #10 is minor axis of ellipse

y = b*np.sin(np.radians(t)) #50 is major axis of ellipse

plt.plot(x,y, label = 'Energ√≠a = $10$, Constante $K$ = $20$ ')
plt.legend(fontsize = 20)
plt.grid()

#Sexta Grafica
E= 10
k= 30

m=5


a = np.sqrt(2*E/k) #Eje mayor de la elipse

b = np.sqrt(2*m*E) #Eje menor de la elipse


t = np.linspace(0,360,360)

x =a*np.cos(np.radians(t)) #10 is minor axis of ellipse

y = b*np.sin(np.radians(t)) #50 is major axis of ellipse

plt.plot(x,y, label = 'Energ√≠a = $10$, Constante $K$ = $30$ ')
plt.legend(fontsize = 20)
plt.grid()




plt.title("Espacio de Fases del Osiclador Arm√≥nico Simple", fontsize = 30)
plt.xlabel("$x$", fontsize = 20)
plt.ylabel("$p$", fontsize = 20)
#plt.xlim(-2.1,4)
plt.savefig("Grafica__1.png", dpi=500)

plt.show()
```
