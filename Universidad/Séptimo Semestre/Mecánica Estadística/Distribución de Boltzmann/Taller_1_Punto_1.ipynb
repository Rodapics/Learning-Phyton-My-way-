```python
A = 1
B = 1
m = 1
E = 50
dt = 0.01
def f(x,A,B):
    V=A*x**2+B*x**4
    return V

x = np.linspace(-3,3)
y = f(x,3,5)
plt.plot(x,y)
plt.title('Potencial $V(x).$', fontsize = 25)
plt.xlabel('Posici√≥n $V(x)$', fontsize = 20)
plt.ylabel('Potencial $(x)$', fontsize = 20)
plt.grid()
plt.savefig("Grafica_2.png", dpi=500)
plt.show()
```

```python

# Definir constantes
A = 1
B = 1
m = 1

dt = 0.01

# Definir funci√≥n de energ√≠a potencial
def V(x):
    return A * x**2 + B**4

# Definir funci√≥n de ecuaci√≥n de movimiento
def dx_dt(x, v):
    return v

def dv_dt(x, v):
    return -2*A*x/m

# Definir condiciones iniciales
x0 = 0
v0 = np.sqrt(2*(E - V(x0))/m)

# Resolver ecuaci√≥n de movimiento num√©ricamente
t = np.arange(0, 10, dt)
x = np.zeros_like(t)
v = np.zeros_like(t)
x[0] = x0
v[0] = v0

for i in range(1, len(t)):
    x[i] = x[i-1] + dx_dt(x[i-1], v[i-1]) * dt
    v[i] = v[i-1] + dv_dt(x[i-1], v[i-1]) * dt

# Crear diagrama de fases
plt.plot(x, v, label ='t=10s')
plt.xlabel('Posici√≥n x', fontsize=20)
plt.ylabel('Velocidad v', fontsize=20)
plt.title('Diagrama de fases', fontsize=25)
plt.legend()
plt.savefig("Grafica__3.png", dpi=500)
plt.show()
```
