```python
#Definimos variables.
k= 1.380649*10**(-23)
mh = 3.32*10**(-27)
mo = 5.31*10**(-26)
mn = 4.65*10**(-26)
t = 300
```

```python
#probar
k
```

```python
#Definimos la función
x=np.linspace(0,5000,10000)
y1= np.e**(-(x*x*mh)/(2*k*t))*x*x*4*np.pi*(mh/(2*np.pi*k*t))**(3/2)
y2= np.e**(-(x*x*mo)/(2*k*t))*x*x*4*np.pi*(mo/(2*np.pi*k*t))**(3/2)
y3= np.e**(-(x*x*mn)/(2*k*t))*x*x*4*np.pi*(mn/(2*np.pi*k*t))**(3/2)
```

```python
#Graficamos:
plt.figure(figsize=(16,12))
plt.plot(x,y1,'-',color='purple',  label='Hidrogeno')
plt.plot(x,y2,'-',color='green', label='Oxigeno')
plt.plot(x,y3,'-',color='red', label='Nitrogeno')
plt.title('Función de distribución de velocidad de Maxwell para diferentes moleculas.')
plt.xlabel('Velocidad')
plt.ylabel('Distribución de Velocidad')
plt.legend(loc='upper right', fontsize=10)  #Colocación de la legend y el tamaño de la fuente usada en este
plt.grid()
plt.show()


```
