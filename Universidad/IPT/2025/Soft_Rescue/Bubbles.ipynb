```python
#Librerias Necesarias
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scienceplots
import pandas as pd
from scipy.stats import linregress
plt.style.reload_library()
plt.style.use('science')
```

# 10 cm

## 1

```python
# Leer el archivo Excel y seleccionar la hoja "Burbuja 1"
df = pd.read_excel("10 cm.xlsx", sheet_name="Burbuja 1")

# Extraer las columnas 'x' y 't'
x = df['x']
t = df['t']
y = df['y']
# Definir los puntos de inicio y fin para el intervalo
t1, t2 = 1.2, 1.4  # Aquí debes definir los valores específicos de t1 y t2
mask = (t >= t1) & (t <= t2)

# Filtrar los datos en el intervalo
t_interval = t[mask]
x_interval = x[mask]
y_interval = y[mask]
# Realizar el ajuste lineal
slope, intercept, r_value, p_value, std_err = linregress(t_interval, y_interval)

# Calcular r^2
r_squared = r_value**2

# Imprimir los resultados
print(f'Pendiente: {slope}')
print(f'Intercepto: {intercept}')
print(f'R^2: {r_squared}')




# Graficar los datos originales y la línea ajustada
plt.figure(figsize=(5, 6))
plt.scatter(t, y, label='Experimental Data', color='green', linewidth=3)
plt.plot(t_interval, slope * t_interval + intercept, label=f'Linearization: y = {slope:.2f}x + {intercept:.2f}', color='r', linewidth=3)
plt.xlabel('Time (t)', fontsize=20, fontweight='bold')
plt.ylabel('Height (y)', fontsize=20, fontweight='bold')
plt.title('Bubble released at 10 cm', fontsize=20, fontweight='bold')
plt.legend(fontsize=12, loc='upper right', frameon=True)
plt.grid(True)
plt.tick_params(axis='both', which='major', labelsize=15)
# Agregar la caja de texto con el valor de r^2
plt.text(0.65, 0.8, f'$v = {slope:.2f} m/s$' , transform=plt.gca().transAxes, fontsize=14, color='black',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))
plt.text(0.7, 0.7, f'$r^2 = {r_squared:.4f}$' , transform=plt.gca().transAxes, fontsize=14, color='black',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.savefig("y_vs_t_10_cm_1", dpi=500)  # Guarda como archivo PNG con alta resolución

plt.show()
```

# Theoretical Model.


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, sqrt, tanh

# Constantes dadas
m = 1.0     # masa (kg)
g = 9.81    # gravedad (m/s^2)
rho = 1.2   # densidad del fluido (kg/m^3)
A = 0.1     # área transversal (m^2)
Cd = 0.5    # coeficiente de arrastre

# Definimos la función de velocidad
k = sqrt((2 * m * g) / (rho * A * Cd))
omega = sqrt((g * rho * A * Cd) / (2 * m))

def velocity(t):
    return k * np.tanh(t * omega)

# Integramos simbólicamente para obtener la posición
# Usamos SymPy para definir x(t) = \int v(t) dt

# Variables simbólicas
t = symbols('t')
vel_sym = k * tanh(t * omega)
pos_sym = integrate(vel_sym, t)

# Convertimos x(t) a una función evaluable
from sympy.utilities.lambdify import lambdify
position = lambdify(t, pos_sym, modules='numpy')

# Valores para el tiempo
t_values = np.linspace(0, 10, 500)

# Calculamos velocidad y posición
t_values_numeric = np.array(t_values)
vel_values = velocity(t_values_numeric)
pos_values = position(t_values_numeric)

# Graficamos velocidad y posición
plt.figure(figsize=(12, 6))

# Gráfica de velocidad
plt.subplot(1, 2, 1)
plt.plot(t_values, vel_values, label='Velocidad $v(t)$', color='blue')
plt.title('Velocidad en función del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.grid()
plt.legend()

# Gráfica de posición
plt.subplot(1, 2, 2)
plt.plot(t_values, pos_values, label='Posición $x(t)$', color='green')
plt.title('Posición en función del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
