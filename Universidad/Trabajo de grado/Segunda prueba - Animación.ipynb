```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from matplotlib.animation import PillowWriter

# Configuraci√≥n de las variables
x = np.linspace(-np.pi, np.pi, 1000)
duration_animation = 2  # Duraci√≥n de la animaci√≥n en segundos
pause_duration = 0.5  # Duraci√≥n de la pausa en segundos
fps = 60  # Fotogramas por segundo

# Funciones de la animaci√≥n
def f1(x, t): 
    return -2 * np.cos(x) + (t - 3)

def f2(x, t): 
    return -2 * np.cos(x) - (t + 3)

def f3(x, t): 
    return -2 * np.cos(x) - 3

# Configuraci√≥n de la figura
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ln1, = plt.plot([], [], label=r'$-cos(x)$', color='green', linewidth=2)
ln2, = plt.plot([], [], label=r'$-cos(x)$', color='red', linewidth=2)
ln3, = plt.plot([], [], label=r'$-cos(x)$', color='blue', linewidth=2)
time_text = ax.text(0.65, 0.85, '', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))

ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-6, 3)

# N√∫mero total de frames para la animaci√≥n y la pausa
frames_animation = int(duration_animation * fps)
frames_pause = int(pause_duration * fps)
total_frames = frames_animation + frames_pause

# Funci√≥n de animaci√≥n
def animate(i):
    if i < frames_animation:
        ln1.set_data(x, f1(x, i / fps))
        ln2.set_data(x, f2(x, i / fps))
        ln3.set_data(x, f3(x, i / fps))
        time_text.set_text('B={:2f}'.format(i / fps))
    else:
        time_text.set_text('B={:2f}'.format(frames_animation / fps))  # Mantener el √∫ltimo valor del tiempo

# Crear y guardar la animaci√≥n
ani = animation.FuncAnimation(fig, animate, frames=total_frames, interval=1000/fps)
ani.save('image223.gif', writer='pillow', fps=fps, dpi=100)
```

## Grafica Funci√≥n Coseno

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuraciones generales para estilo de la gr√°fica
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.size": 14,  # Tama√±o de la fuente general
    "lines.linewidth": 2,
    "axes.linewidth": 1.5
})

# Datos para la funci√≥n coseno
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.cos(x)

# Creaci√≥n de la figura y el eje
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar la funci√≥n coseno
ax.plot(x, y, label=r'$\cos(\hat{\theta}_{q})$', color='b')

# Personalizaci√≥n de los ejes
ax.set_xlabel(r'$\hat{\theta}_{q}$', fontsize=25)  # Tama√±o del t√≠tulo del eje x
ax.set_ylabel(r'$\cos(\hat{\theta}_{q})$', fontsize=25)  # Tama√±o del t√≠tulo del eje y

# Configuraci√≥n de ticks en formato LaTeX
ax.set_xticks([-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
ax.set_xticklabels([r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels([r'$-1$', r'$0$', r'$1$'])

# Ajustar el tama√±o de los ticks en ambos ejes
ax.tick_params(axis='x', labelsize=25)  # Tama√±o de los ticks en el eje x
ax.tick_params(axis='y', labelsize=16)  # Tama√±o de los ticks en el eje y

# Marcar los puntos m√≠nimos con "bolitas"
min_x = [-np.pi, np.pi]  # Puntos x donde el coseno tiene m√≠nimos
min_y = [-1, -1]  # Valor m√≠nimo del coseno
ax.scatter(min_x, min_y, color='g', s=100, zorder=5)  # "Bolitas" en los m√≠nimos

# A√±adir una cuadr√≠cula sutil
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Ajuste del layout
plt.tight_layout()

# Guardar la gr√°fica con alta resoluci√≥n
plt.savefig("coseno_minimos.png", dpi=500)

# Mostrar la gr√°fica
plt.show()
```
