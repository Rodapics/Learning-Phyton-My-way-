```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from matplotlib.animation import PillowWriter

# Configuración de las variables
x = np.linspace(-np.pi, np.pi, 1000)
duration_animation = 2  # Duración de la animación en segundos
pause_duration = 0.5  # Duración de la pausa en segundos
fps = 60  # Fotogramas por segundo

# Funciones de la animación
def f1(x, t): 
    return -2 * np.cos(x) + (t - 3)

def f2(x, t): 
    return -2 * np.cos(x) - (t + 3)

def f3(x, t): 
    return -2 * np.cos(x) - 3

# Configuración de la figura
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ln1, = plt.plot([], [], label=r'$-cos(x)$', color='green', linewidth=2)
ln2, = plt.plot([], [], label=r'$-cos(x)$', color='red', linewidth=2)
ln3, = plt.plot([], [], label=r'$-cos(x)$', color='blue', linewidth=2)
time_text = ax.text(0.65, 0.85, '', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))

ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-6, 3)

# Número total de frames para la animación y la pausa
frames_animation = int(duration_animation * fps)
frames_pause = int(pause_duration * fps)
total_frames = frames_animation + frames_pause

# Función de animación
def animate(i):
    if i < frames_animation:
        ln1.set_data(x, f1(x, i / fps))
        ln2.set_data(x, f2(x, i / fps))
        ln3.set_data(x, f3(x, i / fps))
        time_text.set_text('B={:2f}'.format(i / fps))
    else:
        time_text.set_text('B={:2f}'.format(frames_animation / fps))  # Mantener el último valor del tiempo

# Crear y guardar la animación
ani = animation.FuncAnimation(fig, animate, frames=total_frames, interval=1000/fps)
ani.save('image223.gif', writer='pillow', fps=fps, dpi=100)
```
