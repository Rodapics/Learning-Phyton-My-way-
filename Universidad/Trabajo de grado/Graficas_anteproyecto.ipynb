```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos para las funciones cosenoidales invertidas
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y_blue = -np.cos(x)        # Funcin coseno invertida (azul)
y_red = -np.cos(x) - 1      # Funcin coseno invertida (roja) desplazada hacia abajo
y_green = -np.cos(x) + 1    # Funcin coseno invertida (verde) desplazada hacia arriba

# Crear la grfica
plt.figure(figsize=(8, 6))
plt.plot(x, y_blue, label=r'$-cos(x)$', color='blue', linewidth=2)
plt.plot(x, y_red, label=r'$-cos(x)-1$', color='red', linewidth=2)
plt.plot(x, y_green, label=r'$-cos(x)+1$', color='green', linewidth=2)

# Opciones de la grfica
plt.xlabel('x (radianes)')
plt.ylabel('y')
plt.title('Grfico de funciones $-cos(x)$ desplazadas')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Mostrar la grfica
plt.show()

```
