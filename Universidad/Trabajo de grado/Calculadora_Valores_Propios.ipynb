## Verificación Matriz Original

```python
import sympy as sp

# Definir variables simbólicas
g_0, g_2 = sp.symbols('g_0 g_2')
a, b, c = sp.symbols('a b c')

# Definir una lista de casos (a, b, c) que queremos probar
cases = [
    (1, 1, 1),  # Caso a = b = c
    #(2, 1, 1),  # Caso a = 2b, c = b  # Caso a = 1.5b, c = b
]

# Valores numéricos para g_0 y g_2 (puedes ajustarlos según lo que necesites)
g_0_val = 1.0
g_2_val = 1.0

# Función para calcular los valores propios y vectores propios simbólicamente y numéricamente
def calculate_eigenvalues_and_vectors(a_val, b_val, c_val):
    # Definir relaciones simbólicas entre los parámetros
    beta_squared = b_val / a_val
    alpha_squared = a_val / b_val
    gamma_squared = a_val * b_val / c_val**2
    nu = c_val / a_val + c_val / b_val

    beta = sp.sqrt(beta_squared)
    alpha = sp.sqrt(alpha_squared)
    gamma = sp.sqrt(gamma_squared)

    # Definir la matriz simbólica para el caso específico de a, b, c
    matrix = sp.Matrix([
    [(g_2*(13) + 2*g_0)/12, (g_2*(2) + g_0)/6, (g_2*(1) + g_0*(5))/12],
    [(g_2*(2) + g_0)/6, (g_2*(10) + g_0*(5))/12, (g_2*(2) + g_0)/6],
    [(g_2*(1) + g_0*(5))/12, (g_2*(2) + g_0)/6, (g_2*(13) + 2*g_0)/12]
  ])

    # Calcular valores propios y vectores propios simbólicamente
    eigenvalues = matrix.eigenvals()
    eigenvectors = matrix.eigenvects()

    # Calcular valores y vectores propios numéricos sustituyendo g_0 y g_2
    matrix_numeric = matrix.subs({g_0: g_0_val, g_2: g_2_val})
    eigenvalues_numeric = matrix_numeric.eigenvals()
    eigenvectors_numeric = matrix_numeric.eigenvects()

    return eigenvalues, eigenvectors, eigenvalues_numeric, eigenvectors_numeric

# Iterar sobre cada caso y calcular valores y vectores propios simbólicos y numéricos
for i, (a_val, b_val, c_val) in enumerate(cases):
    eigenvalues, eigenvectors, eigenvalues_numeric, eigenvectors_numeric = calculate_eigenvalues_and_vectors(a_val, b_val, c_val)
    
    print(f"\nCaso {i + 1}: a = {a_val}, b = {b_val}, c = {c_val}")
    print("Valores propios simbólicos:")
    sp.pprint(eigenvalues)
    
    print("\nVectores propios simbólicos:")
    for eig in eigenvectors:
        sp.pprint(eig)

    print("\nValores propios numéricos:")
    sp.pprint(eigenvalues_numeric)
    
    print("\nVectores propios numéricos:")
    for eig in eigenvectors_numeric:
        sp.pprint(eig)
    
    print("-" * 40)
```
