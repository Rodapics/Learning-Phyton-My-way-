```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scienceplots
plt.style.reload_library()
plt.style.use('science')

#Función Parabolica

x = np.linspace(-15, 15, 1000)
y1 = x**2

x2 = np.linspace(-10,10,1000)
y2 = np.full_like(x2, 100)  # Una función constante de valor 3



#Tamaño Grafica
plt.figure(figsize=(9, 10))

# Graficar las funciones.
plt.plot(x, y1, label='x**2', color='black')
plt.plot(x2, y2, label='x**2', color='black')


#Definir y graficar las funciones lineales.
x3 = np.linspace(-15,10,1000)
punto1 = (-10, 100)
punto2 = (0, -70)
m = (punto2[1] - punto1[1]) / (punto2[0] - punto1[0]) 
b = punto1[1] - m * punto1[0]
y3 = m * x3 + b    
plt.plot(x3, y3, color='green', linewidth=2)
#------
x4 = np.linspace(-10,15,1000)
punto1 = (10, 100)
punto2 = (0, -70)
m = (punto2[1] - punto1[1]) / (punto2[0] - punto1[0]) 
b = punto1[1] - m * punto1[0]
y3 = m * x4 + b    
plt.plot(x4, y3, color='green', linewidth=2)


# Graficar la función coseno
#plt.plot(x, y2, label='coseno(x)', color='red')

# Dibujar línea para el eje x
plt.axhline(0, color='black', linewidth=2, linestyle='-', xmax=1)

# Dibujar línea para el eje y en la mitad de la gráfica
y_axis_line_position = (plt.ylim()[0] + plt.ylim()[1]) / 2
plt.axvline(0, color='black', linewidth=2, linestyle='-', ymax=1)

# Agregar bolas rellenas en puntos específicos
#points_x = [1.5, -3, 4]
#points_y = [0.5, -0.7, 0.2]
#plt.scatter(points_x, points_y, color='green', s=50, alpha=0.7)




# Agregar líneas horizontales ajustables
lines_start_y = [81]
lines_end_y = [64]
for start_y, end_y in zip(lines_start_y, lines_end_y):
    plt.hlines(y=start_y, xmin=-9, xmax=9, color='black', linewidth=1, linestyle='-')
    plt.hlines(y=end_y, xmin=-8, xmax=8, color='black', linewidth=1, linestyle='-')
#----
lines_start_y = [49]
lines_end_y = [36]
for start_y, end_y in zip(lines_start_y, lines_end_y):
    plt.hlines(y=start_y, xmin=-7, xmax=7, color='black', linewidth=1, linestyle='-')
    plt.hlines(y=end_y, xmin=-6, xmax=6, color='black', linewidth=1, linestyle='-')


#----
lines_start_y = [25]
lines_end_y = [16]
for start_y, end_y in zip(lines_start_y, lines_end_y):
    plt.hlines(y=start_y, xmin=-5, xmax=5, color='black', linewidth=1, linestyle='-')
    plt.hlines(y=end_y, xmin=-4, xmax=4, color='black', linewidth=1, linestyle='-')
#----
lines_start_y = [9]
lines_end_y = [4]
for start_y, end_y in zip(lines_start_y, lines_end_y):
    plt.hlines(y=start_y, xmin=-3, xmax=3, color='black', linewidth=1, linestyle='-')
    plt.hlines(y=end_y, xmin=-2, xmax=2, color='black', linewidth=1, linestyle='-')
    
    
    
    
# Agregar líneas verticales ajustables
lines_start_x = [-11]
lines_end_x = [11]
for start_x, end_x in zip(lines_start_x, lines_end_x):
    plt.vlines(x=start_x, ymin=-1, ymax=121, color='#0000FF', linewidth=2, linestyle='dashed')
    plt.vlines(x=end_x, ymin=-1, ymax=121, color='#0000FF', linewidth=2, linestyle='dashed')
#-----
lines_start_x = [-9]
lines_end_x = [9]
for start_x, end_x in zip(lines_start_x, lines_end_x):
    plt.vlines(x=start_x, ymin=-1, ymax=81, color='#0000FF', linewidth=2, linestyle='dashed')
    plt.vlines(x=end_x, ymin=-1, ymax=81, color='#0000FF', linewidth=2, linestyle='dashed')
#-----
lines_start_x = [-10]
lines_end_x = [10]
for start_x, end_x in zip(lines_start_x, lines_end_x):
    plt.vlines(x=start_x, ymin=-1, ymax=100, color='gray', linewidth=2, linestyle='-')
    plt.vlines(x=end_x, ymin=-1, ymax=100, color='gray', linewidth=2, linestyle='-')

    
    
    
    
    
# Agregar letras en formato LaTeX (Gamma)
gamma_positions = [-10, 10]
for position in gamma_positions:
    plt.text(position, 120, r'$\Lambda$', fontsize=25, ha='center', va='center', color='#0000FF')


    
    
# Configurar ticks personalizados en el eje x y asignar nombres especiales en formato LaTeX
ct = 10
epsilon = 1

custom_ticks = [-epsilon-ct, -ct, -ct+epsilon+0.2, -epsilon+ct, ct, ct+epsilon ]

tick_labels = [r'$($',r'$-k_f$ ',r'$)$','(',r'$k_f$',')']
plt.xticks(custom_ticks, labels=tick_labels, fontsize=18)


#Rellenar la grafica
plt.fill_between(x,y2,y1, where=(y2 > y1),interpolate=True, color='#008cff', alpha=0.1, label='Área Rellena')



# Ocultar etiquetas en el eje y
plt.yticks([])

#Colocar los títulos de las graficas.
plt.text(1, 150, r'$\epsilon$', fontsize=30, ha='center', va='center', color='black')
plt.text(15, -8, r'$k$', fontsize=25, ha='center', va='center', color='black')
#Colocar Energía de Fermi
plt.text(8, 108, r'$E_f$', fontsize=30, ha='center', va='center', color='red')

plt.ylim(-80,160)
# Mostrar el gráfico

# Configurar los ticks dentro de la gráfica
plt.tick_params(axis='both', which='both', direction='in', top=True, right=True)
plt.savefig('nosniztion.png', dpi=500)

plt.show()
```
