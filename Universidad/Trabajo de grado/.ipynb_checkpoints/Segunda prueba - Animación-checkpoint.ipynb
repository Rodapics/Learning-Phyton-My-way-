```python
from celluloid import Camera
import matplotlib.pyplot as plt
from IPython.display import HTML
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scienceplots


theta = np.linspace(0, 4 * np.pi, 100)

fig = plt.figure()
camera = Camera(fig)

for i in range(100):
  theta_i = theta[0:i]
  radius_i = theta_i
  plt.polar(theta_i, radius_i)
  camera.snap()

animation = camera.animate(interval=50, blit=True)

animation.save('my_animation.gif')
```

```python
#Create the animation
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import animation
from matplotlib.animation import PillowWriter


def f(x,t): #Permite definir una función de dos variables.
  return np.sin(x-3*t)

x = np.linspace(0,10*np.pi,1000)




fig, ax =plt.subplots(1,1,figsize=(10,6))
ln1, = plt.plot([],[]) #Se coloca grafica vacía
time_text = ax.text(0.65,0.95, '', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black')) #Se crea la caja de tiempo vacía
ax.set_xlim(0,10*np.pi)
ax.set_ylim(-1.5,1.5) #Límites


#Creamos la función de animación. Frame number i.
def animate(i):
  ln1.set_data(x,f(x,1/50*i)) #Se le agregan los datos
  time_text.set_text('t={:2f}'.format(i/50)) #Se crea la variable de tiempo que se va a mostrar

ani = animation.FuncAnimation(fig, animate, frames=240,interval=50) #Parametros de la animación
ani.save('image.gif', writer='pillow',fps=50,dpi=1000) #Para guardar
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from matplotlib.animation import PillowWriter

x = np.linspace(-np.pi, np.pi, 1000)

def f1(x, t):  # Permite definir una función de dos variables. Para SIGMA = 1
    return -2 * np.cos(x) + (t - 3)

def f2(x, t):  # Permite definir una función de dos variables. PARA SIGMA = -1
    return -2 * np.cos(x) - (t + 3)

def f3(x, t):  # Permite definir una función de dos variables. PARA SIGMA = 0
    return -2 * np.cos(x) - 3

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ln1, = plt.plot([], [], label=r'$-cos(x)$', color='green', linewidth=2)  # Se coloca gráfica vacía
ln2, = plt.plot([], [], label=r'$-cos(x)$', color='red', linewidth=2)  # Se coloca gráfica vacía
ln3, = plt.plot([], [], label=r'$-cos(x)$', color='blue', linewidth=2)  # Se coloca gráfica vacía
time_text = ax.text(0.65, 0.85, '', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))  # Se crea la caja de tiempo vacía

ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-6, 3)  # Límites

# Creamos la función de animación. Frame number i.
def animate(i):
    if i < 120:
        ln1.set_data(x, f1(x, 1/60 * i))  # Se le agregan los datos
        ln2.set_data(x, f2(x, 1/60 * i))  # Se le agregan los datos
        ln3.set_data(x, f3(x, 1/60 * i))  # Se le agregan los datos
        time_text.set_text('B={:2f}'.format(i/60))  # Se crea la variable de tiempo que se va a mostrar
    else:
        time_text.set_text('B={:2f}'.format(2))  # Mantener el último valor del tiempo

# Número total de frames: 120 frames de animación + 30 frames para la pausa (medio segundo a 60 FPS)
total_frames = 120 + 60

ani = animation.FuncAnimation(fig, animate, frames=total_frames)  # Parámetros de la animación

ani.save('image2222.gif', writer='pillow', fps=60, dpi=100)  # Para guardar
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from matplotlib.animation import PillowWriter

# Configuración de las variables
x = np.linspace(-np.pi, np.pi, 1000)
duration_animation = 2  # Duración de la animación en segundos
pause_duration = 0.5  # Duración de la pausa en segundos
fps = 60  # Fotogramas por segundo

# Funciones de la animación
def f1(x, t): 
    return -2 * np.cos(x) + (t - 3)

def f2(x, t): 
    return -2 * np.cos(x) - (t + 3)

def f3(x, t): 
    return -2 * np.cos(x) - 3

# Configuración de la figura
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ln1, = plt.plot([], [], label=r'$-cos(x)$', color='green', linewidth=2)
ln2, = plt.plot([], [], label=r'$-cos(x)$', color='red', linewidth=2)
ln3, = plt.plot([], [], label=r'$-cos(x)$', color='blue', linewidth=2)
time_text = ax.text(0.65, 0.85, '', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))

ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-6, 3)

# Número total de frames para la animación y la pausa
frames_animation = int(duration_animation * fps)
frames_pause = int(pause_duration * fps)
total_frames = frames_animation + frames_pause

# Función de animación
def animate(i):
    if i < frames_animation:
        ln1.set_data(x, f1(x, i / fps))
        ln2.set_data(x, f2(x, i / fps))
        ln3.set_data(x, f3(x, i / fps))
        time_text.set_text('B={:2f}'.format(i / fps))
    else:
        time_text.set_text('B={:2f}'.format(frames_animation / fps))  # Mantener el último valor del tiempo

# Crear y guardar la animación
ani = animation.FuncAnimation(fig, animate, frames=total_frames, interval=1000/fps)
ani.save('image223.gif', writer='pillow', fps=fps, dpi=100)
```

## Animación Completa HD

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from matplotlib.animation import PillowWriter
import scienceplots
plt.style.reload_library()
plt.style.use('science')

# Variable espacio-temporal.
x = np.linspace(-np.pi, np.pi, 1000)



#Variables de tiempo de la animación
duration_animation = 1  # Duración de la animación en segundos
pause_duration = 4  # Duración de la pausa en segundos
fps = 60  # Fotogramas por segundo
interval = 1000 / fps  # Intervalo en milisegundos



# Funciones de la animación
def f1(x, t):
    return -2 * np.cos(x) + (t - 1) #Función de Sigma = 1

def f2(x, t):
    return -2 * np.cos(x) - (t + 1) #Función de Sigma = -1

def f3(x, t):
    return -2 * np.cos(x) - 1       #Función de Sigma = 0



# Configuración de la figura
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ln1, = plt.plot([], [], label=r'$\epsilon_\sigma= 1$', color='green', linewidth=2)
ln2, = plt.plot([], [], label=r'$\epsilon_\sigma=-1$', color='red', linewidth=2)
ln3, = plt.plot([], [], label=r'$\epsilon_\sigma= 0$', color='blue', linewidth=1, linestyle='--')
time_text = ax.text(0.8, 0.25, '', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))


#Límites
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-4, 2)


# Títulos y leyenda
#ax.set_title('Animación de Funciones Trigonométricas', fontsize=16)  # Título del gráfico
# Títulos de los ejes en LaTeX
ax.set_xlabel(r'$k$', fontsize = 25)  # Título del eje X con LaTeX
ax.set_ylabel(r'$\epsilon_\sigma (k) $', fontsize = 25)  # Título del eje Y con LaTeX

ax.legend(loc='lower right', fontsize = 15)  # Mostrar la leyenda en la esquina superior izquierda

# Insertar texto con LaTeX en la figura
#ax.text(-np.pi, 2, r'$\sigma$', fontsize=18, color='purple')  # Texto en la esquina superior izquierda


# Colocar ticks específicos en el eje X
xticks = [-np.pi, np.pi]
xticklabels = [r'$-\pi/a$', r'$\pi/a$']
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, fontsize=20)



# Ocultar los números de los ejes
#ax.set_xticks([])  # Ocultar los números en el eje X
#ax.set_yticks([])  # Ocultar los números en el eje Y


# Número total de frames para la animación y la pausa
frames_animation = int(duration_animation * fps)
frames_pause = int(pause_duration * fps)
total_frames = frames_animation + frames_pause

# Función de animación
def animate(i):
    if i < frames_animation:
        ln1.set_data(x, f1(x, i / fps))
        ln2.set_data(x, f2(x, i / fps))
        ln3.set_data(x, f3(x, i / fps))
        time=i / fps
        time_text.set_text('B={:.2f}'.format(time))
    else:
        ln1.set_data(x, f1(x, 1))
        ln2.set_data(x, f2(x, 1))
        ln3.set_data(x, f3(x, 1))        
        time_text.set_text('B=1.00')  # Mantener el último valor del tiempo

# Crear y guardar la animación
ani = animation.FuncAnimation(fig, animate, frames=total_frames, interval=interval)
ani.save('v1.gif', writer='pillow', fps=fps, dpi=500)
```

```python
#Primera, B=1

# Variable espacio-temporal.
x = np.linspace(-np.pi, np.pi, 1000)


# Funciones de la animación
def f1(x, t):
    return -2 * np.cos(x) + (t - 1) #Función de Sigma = 1

def f2(x, t):
    return -2 * np.cos(x) - (t + 1) #Función de Sigma = -1

def f3(x, t):
    return -2 * np.cos(x) - 1       #Función de Sigma = 0

fig, ax = plt.subplots(1, 1, figsize=(8, 6))


plt.plot(x, f1(x,1), label=r'$\epsilon_\sigma= 1$', color='green', linewidth=2)
plt.plot(x, f2(x,1), label=r'$\epsilon_\sigma=-1$', color='red', linewidth=2)
plt.plot(x, f3(x,0), label=r'$\epsilon_\sigma= 0$', color='blue', linewidth=1, linestyle='--')

#Límites
plt.xlim(-np.pi, np.pi)
plt.ylim(-4, 2)


#Títulos y leyenda.
plt.xlabel(r'$k$', fontsize = 25)
plt.ylabel(r'$\epsilon_\sigma (k) $', fontsize = 25)

plt.legend(loc='lower right', fontsize = 15)


# Colocar ticks específicos en el eje X
xticks = [-np.pi, np.pi]
xticklabels = [r'$-\pi/a$', r'$\pi/a$']

time_text = ax.text(0.8, 0.25, 'B=1.00', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))


plt.xticks(xticks,xticklabels, fontsize=20)

plt.show()
```
