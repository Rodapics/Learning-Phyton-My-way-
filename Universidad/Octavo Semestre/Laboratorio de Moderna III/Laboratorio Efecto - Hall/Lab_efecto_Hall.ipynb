```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
plt.style.reload_library()
plt.style.use(['science','notebook'])
from itertools import combinations
print('done')
```

```python
Arr = pd.read_csv("I_0_09.csv")



def lin(x, a, b):
    return a*x +b


res, cov = curve_fit(lin, Arr["V_B"], Arr["V_A"])



x= np.arange(0,1.7, 0.01)


plt.plot(x, lin(x, res[0], res[1]), 'r-', label='Ajuste Lineal')
plt.plot(Arr["V_B"], Arr["V_A"], 'o', label='Datos')


plt.legend()
plt.grid(True)
plt.xlabel(r'$B$ (mT)')
plt.ylabel( r' $V$ (v)')

plt.show()
```
