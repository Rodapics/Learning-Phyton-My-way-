```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definir la función lineal
def lin(x, a, b):
    return a * x + b

# Generar datos de ejemplo
np.random.seed(0)  # Fijar la semilla para reproducibilidad
x_data = np.linspace(0, 10, 100)
y_data = 2.5 * x_data + 1.2 + np.random.normal(scale=2, size=len(x_data))

# Ajustar la función lineal a los datos
params, covariance = curve_fit(lin, x_data, y_data)

# Obtener los parámetros del ajuste
a_fit, b_fit = params

# Imprimir los parámetros del ajuste
print("Parámetros del ajuste:")
print("a =", a_fit)
print("b =", b_fit)

# Generar datos ajustados usando la función lineal ajustada
y_fit = lin(x_data, a_fit, b_fit)

# Graficar los datos y la función ajustada
plt.scatter(x_data, y_data, label='Datos')
plt.plot(x_data, y_fit, color='red', label='Ajuste lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
