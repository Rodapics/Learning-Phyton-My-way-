## Ln(I) vs V

```python
def lin(x, a, b):
    return a*x +b


y1= np.log(t1['a'].iloc[10:])
y2= np.log(t2['a'].iloc[4:])
y3= np.log(t3['a'].iloc[3:])
y4= np.log(t4['a'].iloc[4:])
y5= np.log(t5['a'].iloc[6:])
y6= np.log(t6['a'].iloc[7:])
y7= np.log(t7['a'].iloc[8:])
y8= np.log(t8['a'].iloc[3:])
y9= np.log(t9['a'].iloc[7:])
y10= np.log(t10['a'].iloc[6:])


res1, cov1 = curve_fit(lin, t1["v"].iloc[10:], y1)
res2, cov2 = curve_fit(lin, t2["v"].iloc[4:], y2)
res3, cov3 = curve_fit(lin, t3["v"].iloc[3:], y3)
res4, cov4 = curve_fit(lin, t4["v"].iloc[4:], y4)
res5, cov5 = curve_fit(lin, t5["v"].iloc[6:], y5)
res6, cov6 = curve_fit(lin, t6["v"].iloc[7:], y6)
res7, cov7 = curve_fit(lin, t7["v"].iloc[8:], y7)
res8, cov8 = curve_fit(lin, t8["v"].iloc[3:], y8)
res9, cov9 = curve_fit(lin, t9["v"].iloc[7:], y9)
res10, cov10 = curve_fit(lin, t10["v"].iloc[6:], y10)

x= np.arange(0.35,0.92, 0.01)

plt.plot(x, lin(x, res1[0], res1[1]), '-')
plt.plot(x, lin(x, res2[0], res2[1]), '-')
plt.plot(x, lin(x, res3[0], res3[1]), '-')
plt.plot(x, lin(x, res4[0], res4[1]), '-')
plt.plot(x, lin(x, res5[0], res5[1]), '-')
plt.plot(x, lin(x, res6[0], res6[1]), '-')
plt.plot(x, lin(x, res7[0], res7[1]), '-')
plt.plot(x, lin(x, res8[0], res8[1]), '-')
plt.plot(x, lin(x, res9[0], res9[1]), '-')
plt.plot(x, lin(x, res10[0], res10[1]), '-')




plt.plot(t1["v"].iloc[10:], y1, 'o', label="295 K")
plt.plot(t2["v"].iloc[4:], y2, 'o', label="323 K")
plt.plot(t3["v"].iloc[3:], y3, 'o', label="338 K")
plt.plot(t4["v"].iloc[4:], y4, 'o', label="353 K")
plt.plot(t5["v"].iloc[6:], y5, 'o', label="368 K")
plt.plot(t6["v"].iloc[7:], y6, 'o', label="383 K")
plt.plot(t7["v"].iloc[8:], y7, 'o', label="398 K")
plt.plot(t8["v"].iloc[3:], y8, 'o', label="413 K")
plt.plot(t9["v"].iloc[7:], y9, 'o', label="428 K")
plt.plot(t10["v"].iloc[6:], y10, 'o', label="443 K")

plt.legend()
plt.xlabel(r' Voltaje $(V)$', fontsize = 20)
plt.ylabel(r'Ln (I)', fontsize = 20)
plt.title('Corriente (A) en función de Voltaje (V) a diferentes temperaturas', fontsize = 20)
plt.savefig('gap2.png', dpi=100) 
plt.xlim(0.3,1.2)
#plt.ylim(-10,2)

plt.grid(True)

plt.show()
```
