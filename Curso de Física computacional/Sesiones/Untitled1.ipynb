```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para las líneas
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear la gráfica principal
plt.figure(figsize=(8, 6))

# Graficar las líneas
plt.plot(x, y1, label='Línea 1', color='blue')
plt.plot(x, y2, label='Línea 2', color='red')

# Rellenar el área entre las líneas
plt.fill_between(x, y1, y2, where=(y1 > y2), interpolate=True, color='gray', alpha=0.3, label='Área Rellena')

# Añadir leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_recta(punto1, punto2):
    # Calcular la pendiente (m)
    m = (punto2[1] - punto1[1]) / (punto2[0] - punto1[0])
    
    # Calcular el término independiente (b)
    b = punto1[1] - m * punto1[0]
    
    return m, b

def graficar_recta(punto1, punto2, m, b):
    x = np.linspace(min(punto1[0], punto2[0]), max(punto1[0], punto2[0]), 100)
    y = m * x + b
    
    plt.plot(x, y, label=f'Recta: y = {m:.2f}x + {b:.2f}', color='green')
    #plt.scatter([punto1[0], punto2[0]], [punto1[1], punto2[1]], color='red', marker='o', label='Puntos dados')
    plt.title('Gráfico de la Recta')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.legend()
    plt.grid(True)
    plt.show()

# Definir dos puntos
punto1 = (-9, 81)
punto2 = (0, -40)

# Calcular la ecuación de la recta
pendiente, termino_independiente = calcular_recta(punto1, punto2)

# Graficar la recta
graficar_recta(punto1, punto2, pendiente, termino_independiente)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_recta(punto1, punto2):
    m = (punto2[1] - punto1[1]) / (punto2[0] - punto1[0])
    b = punto1[1] - m * punto1[0]
    return m, b

def graficar_recta(punto1, punto2, m, b):
    x = np.linspace(min(punto1[0], punto2[0]), max(punto1[0], punto2[0]), 100)
    y = m * x + b
    
    plt.plot(x, y, label=f'Recta: y = {m:.2f}x + {b:.2f}')
    plt.scatter([punto1[0], punto2[0]], [punto1[1], punto2[1]], color='red', marker='o', label='Puntos dados')

    # Configurar los ticks dentro de la gráfica
    plt.tick_params(axis='both', which='both', direction='in', top=True, right=True)

    # Ajustar el rango de los ejes para asegurarse de que los ticks estén dentro
    plt.xlim(min(x), max(x))
    plt.ylim(min(y), max(y))

    plt.title('Gráfico de la Recta')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.legend()
    plt.grid(True)
    plt.show()

# Definir dos puntos
punto1 = (-10, 100)
punto2 = (0, -80)

# Calcular la ecuación de la recta
pendiente, termino_independiente = calcular_recta(punto1, punto2)

# Graficar la recta
graficar_recta(punto1, punto2, pendiente, termino_independiente)
```
