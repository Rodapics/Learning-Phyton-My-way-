```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere
print('Done')
```

### The case |00>

Lo primero que veremos es que tenemos un estado $\left|00\right\rangle$, después le aplicamos una compuerta Haddamar y después le aplicaremos una compuerta CX

$$
\left|00\right\rangle
    \xrightarrow{H} \frac{1}{\sqrt{2}}\left(\left|00\right\rangle + \left|10\right\rangle\right)
     \xrightarrow{CX} \frac{1}{\sqrt{2}}\left(\left|00\right\rangle + \left|11\right\rangle\right)
$$

```python
#Definimos el estado incial mencioado.
sv = Statevector.from_label("00")  #Estamos definiendo un vector de estado.
plot_state_qsphere(sv.data)
```

```python
#Ahora definimos y pintamos el circuito cuántico (compuerta Haddamard)
mycircuit1 = QuantumCircuit(2)  #Inicializamos el circuito con el número de bits requerido
mycircuit1.h(0)  #Crea una compuerta Haddamard en el bit 0
#mycircuit1.cx(0, 1) #(control, actua)  #Crea una compuerta controlada C - X donde el qbit de control se encuentra en el primer qubit. 
mycircuit1.draw("mpl")  #Pinta la compuerta
```

```python
#Ahora definimos y pintamos el circuito cuántico (compuerta Haddamard)
mycircuit = QuantumCircuit(2)  #Inicializamos el circuito con el número de bits requerido
mycircuit.h(0)  #Crea una compuerta Haddamard en el bit 0
mycircuit.cx(0, 1) #(control, actua)  #Crea una compuerta controlada C - X donde el qbit de control se encuentra en el primer qubit. 
mycircuit.draw("mpl")  #Pinta la compuerta
```

```python
#Aqui aplicamos que el qbit preaviamente definido, atraviese el circuito cuántico que habiamos definido.
#Pero primero que atraviese solo la Haddamar.


new_sv1 = sv.evolve(mycircuit1) #Aplicamos al estado el circuito.
print(new_sv1)
plot_state_qsphere(new_sv1.data)
```

```python
#Aqui ya atraviesa todo el circuito completo.
new_sv = sv.evolve(mycircuit) #Aplicamos al estado el circuito.
print(new_sv)
plot_state_qsphere(new_sv.data)
```
